```python
# Question 10: Remove all vowels from "Computer Science"
print("\nQuestion 10: Remove all vowels from 'Computer Science'")

text = 'Computer Science'
vowels = "aeiouAEIOU"
result = ""

for ch in text:
    if ch not in vowels:
        result = result + ch

print(result)
```
